# Distancias

```python
import pandas as pd
from scipy.spatial import distance_matrix
```

```python
data = pd.read_csv("datasets/movies/movies.csv", sep=";")
data
```

```python
movies = data.columns.values.tolist()[1:] #[1:] desde la columna uno hacia delante
movies
```

```python
dd1 = distance_matrix(data[movies], data[movies], p=1)##Calcular matrix de distancias
dd2 = distance_matrix(data[movies], data[movies], p=2)
dd10 = distance_matrix(data[movies], data[movies], p=10)
```

```python
def dm_to_df(dd, col_name):  #matrix de distancias hacia data frame
    import pandas as pd
    return pd.DataFrame(dd, index=col_name, columns=col_name)
```

```python
dm_to_df(dd1, data["user_id"])
```

```python
dm_to_df(dd2, data["user_id"])
```

```python
dm_to_df(dd10, data["user_id"])
```

# Clustering jerrquico

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
movies
```

```python
data[movies]
```

```python
Z = linkage(data[movies], "ward") ## ward -> Minimiza los cuadrados de las varianzas
Z ## Muestra las uniones de las variables de acuerdo a la distancia mnima, muestra distancia
##y la cantidad de elementos que hay en el cluster
```

```python
plt.figure(figsize=(25,10))
plt.title("Dendograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10)
plt.show()
```

```python
Z = linkage(data[movies], "average") ## average -> Segn el promedio
Z 
plt.figure(figsize=(25,10))
plt.title("Dendograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10)
plt.show()
```

```python
Z = linkage(data[movies], "complete") ## average -> Segn el promedio
Z 
plt.figure(figsize=(25,10))
plt.title("Dendograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10)
plt.show()
```

```python
Z = linkage(data[movies], "single") ## average -> Segn el promedio
Z 
plt.figure(figsize=(25,10))
plt.title("Dendograma jerrquico para el Clustering")
plt.xlabel("ID de los usuarios de Netflix")
plt.ylabel("Distancia")
dendrogram(Z, leaf_rotation=90., leaf_font_size=10)
plt.show()
```
