# Agregación de datos por Categoría

```python
import numpy as np
import pandas as pd
```

```python
gender = ["Male", "Female"]
income = ["Poor", "Middle Class", "Rich"]
```

```python
n = 500

gender_data = []
income_data = []

for i in range(0,500):
    gender_data.append(np.random.choice(gender))
    income_data.append(np.random.choice(income))
```

```python
gender_data[1:10]
```

```python
income_data[1:10]
```

```python
#Z -> N(0,1) 
#N(m, s) -> m + s * z //s-> desviación estandar
height = 160 + 30 * np.random.randn(n)
weight = 55 + 25 * np.random.randn(n)
age = 30 + 12 * np.random.randn(n)
income = 18000 + 3500 *np.random.randn(n)
```

```python
data = pd.DataFrame(
    {
        "Gender" : gender_data,
        "Economic Status" : income_data,
        "Height" : height,
        "Weight" : weight,
        "Age" : age,
        "Income" : income
    }
)
```

```python
data.head()
```

## Agrupación de datos

```python
grouped_gender = data.groupby("Gender")
```

```python
grouped_gender.groups
```

```python
for names, groups in grouped_gender:
    print(names)
    print(groups)
```

```python
grouped_gender.get_group("Female")
```

```python
double_group = data.groupby(["Gender", "Economic Status"])
```

```python
len(double_group) #Hay seis categoías de agrupación
```

```python
double_group.groups
```

```python
for names, groups in double_group:
    print(names)
    print(groups)
```

## Filtrado de datos

```python
double_group["Age"].filter(lambda x: x.sum()>2400)
```

## Transformación de Vaiables

```python
zscore = lambda x :(x - x.mean())/x.std()
```

```python
z_group = double_group.transform(zscore)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(z_group["Age"])
```

```python
fill_na_mean = lambda x : x.fillna(x.mean()) #Reemplaza cualquier valor NA por la media
```

```python
double_group.transform(fill_na_mean)
```
